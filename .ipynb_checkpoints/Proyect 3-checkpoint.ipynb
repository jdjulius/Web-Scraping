# Step 7 - Loop

```python
name = []
displayName = []
location = []
url_user = []
username = []
hasPremiumAccess = []
isProfileOwner = []
premium = []
views = []
appreciations = []
comments = []
url = []


for result in json:
    name.append(result['name'])
    displayName.append(result['owners'][0]['displayName'])
    location.append(result['owners'][0]['location'])
    url_user.append(result['owners'][0]['url'])
    username.append(result['owners'][0]['username'])
    hasPremiumAccess.append(result['owners'][0]['hasPremiumAccess'])
    isProfileOwner.append(result['owners'][0]['isProfileOwner'])
    premium.append(result['premium'])
    views.append(result['stats']['views']['all'])
    appreciations.append(result['stats']['appreciations']['all'])
    comments.append(result['stats']['comments']['all'])
    url.append(result['url'])
```

# Step 8 - pandas

```python
products = pd.DataFrame({
    'name':name,
    'displayName':displayName,
    'url_user':url_user,
    'username':username,
    'hasPremiumAccess':hasPremiumAccess,
    'isProfileOwner':isProfileOwner,
    'premium':premium,
    'views':views,
    'appreciations':appreciations,
    'comments':comments,
    'url':url
})
```
