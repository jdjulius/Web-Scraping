# Step 1 - Imports

```python
import requests
import pandas as pd
```

# Step 2 - Parameters

```python
params = {
    'ResultOrder': 'Relevance',
    'Term': 'computers',
    'CountryCode': 'GT',
}
```

# Step 3 - Response

```python
response = requests.get('https://www.ebooks.com/api/search/', params=params)
```

# Step 6 - Loop 

```python
title = []
subtitle = []
author = []
publisher = []
publisher_year = []
price = []

for result in resultsJson:
    title.append(result['title'])
    subtitle.append(result['subtitle'])
    author.append(result['authors'][0]['author_name'])
    publisher.append(result['publisher'])
    publisher_year.append(result['publication_year'])
    price.append(result['price'])
    
```

```python
title
```

# Step 7 - pandas dataframe

```python
book_df = pd.DataFrame({'Title':title,
                        'Subtitle': subtitle,
                        'author': author,
                        'publisher': publisher,
                        'publication year': publisher_year,
                        'price': price})
```

```python
book_df
```

```python
book_df.to_excel('books.xlsx', index=False)
```
