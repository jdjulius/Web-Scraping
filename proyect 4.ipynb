```python
make = []
model = []
mileage = []
year = []
engine = []
fuel_type = []
style = []
transmission = []
color = []
price = []

for i in range(1,333):

    headers = {
        'authority': 'www.truecar.com',
        'accept': 'application/json, text/plain, */*',
        'accept-language': 'es,es-GT;q=0.9,en-US;q=0.8,en;q=0.7,es-419;q=0.6',
        'if-none-match': 'W/^\\^53d34bf9837ffe025dd9eaaae373b09d^\\^',
        'referer': 'https://www.truecar.com/used-cars-for-sale/listings/location-chester-ma/?page=' + str(i),
        'sec-ch-ua': '^\\^.Not/A)Brand^\\^;v=^\\^99^\\^, ^\\^Google',
        'sec-ch-ua-mobile': '?0',
        'sec-ch-ua-platform': '^\\^Windows^\\^',
        'sec-fetch-dest': 'empty',
        'sec-fetch-mode': 'cors',
        'sec-fetch-site': 'same-origin',
        'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/103.0.0.0 Safari/537.36',
    }

    params = (
        ('city', 'chester'),
        ('collapse', 'true'),
        ('fallback', 'true'),
        ('include_incentives', 'true'),
        ('include_seo^/[^/]', ['canonicalize_body_styles', 'years', 'inventory_summary', 'faqs']),
        ('include_targeted_incentives', 'true'),
        ('new_or_used', 'u'),
        ('per_page', '30'),
        ('postal_code', '01011'),
        ('search_event', 'true'),
        ('sort^/[^/]', 'best_match'),
        ('sponsored', 'true'),
        ('state', 'ma'),
    )

    response = requests.get('https://www.truecar.com/abp/api/vehicles/used/listings', headers=headers, params=params)

    json = response.json()['listings']

    for result in json:
        make.append(result['vehicle']['make'])
        model.append(result['vehicle']['model'])
        mileage.append(result['vehicle']['mileage'])
        year.append(result['vehicle']['year'])
        engine.append(result['vehicle']['engine'])
        fuel_type.append(result['vehicle']['fuel_type'])
        style.append(result['vehicle']['style'])
        transmission.append(result['vehicle']['transmission'])
        color.append(result['vehicle']['exterior_color_generic'])
        price.append(result['pricing']['list_price'])
```

```python
vehicle_lst = pd.DataFrame({
    'make':make,
    'model':model,
    'mileage':mileage,
    'year':year,
    'engine':engine,
    'fuel_type':fuel_type,
    'style':style,
    'transmission':transmission,
    'color':color,
    'price':price
})
```
